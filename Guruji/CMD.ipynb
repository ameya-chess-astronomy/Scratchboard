```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('../Downloads/m5.txt', usecols=(2,8,14,20,26,32))
```

```python
data = data[np.where(data[:,-1]>95)[0]]
```

```python
ind = np.where(data[:,:-1]<0)
data1 = np.delete(data,ind,axis=0)
```

```python
# ind = np.where(data1[:,:-1]>22)
# data1 = np.delete(data1,ind,axis=0)
```

```python
uv,u,b,v,i = data1[:,:-1].T
```

```python
mag = b
col = b-v
```

```python
s = 4e6* 10**(-b/2.5)
```

```python
plt.figure(figsize=(6,6))
plt.style.use('dark_background')
plt.xkcd()
cm = plt.cm.get_cmap('Spectral_r')
plt.scatter(col, mag, s=s,c=col, cmap=cm, vmin=0, vmax=1.5)
plt.ylim(24,12)
plt.xlim(-0.5,2.5)
plt.title('Colour-Magnitude Diagram for Messier 5', fontsize=14, fontweight='bold', color='k')
plt.xlabel('Colour (B-V)',fontsize=14, fontweight='heavy', color='k')
plt.ylabel('Magnitude (V)',fontsize=14, fontweight='heavy', color='k')
ax = plt.gca()
ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
ax.tick_params(which='both',axis='both',direction='in',width=2,size=0,right=True,top=True)
for tick in ax.xaxis.get_major_ticks():
    tick.label1.set_fontsize(14)
    tick.label1.set_fontweight('bold')
    tick.label1.set_color('black')

for tick in ax.yaxis.get_major_ticks():
    tick.label1.set_fontsize(14)
    tick.label1.set_fontweight('bold')
    tick.label1.set_color('black')

# plt.savefig('Messier.jpg')
```
