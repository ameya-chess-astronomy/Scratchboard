```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
planet = "Venus"
```

```python
dict_planets = {"Mercury":0.39, "Venus":0.72, "Mars":1.52, "Jupiter": 5.2, "Saturn":9.58}
```

```python
def position(t, P, a):
    theta = 2*np.pi/P*t
    x = a* np.cos(theta)
    y = a* np.sin(theta)
    return x, y
```

```python
def do_stuff():
    x0, y0 = position(t, 1/P_earth, a_earth)
    x, y = position(t, 1/P_planet, a_planet)
    
    plt.style.use('dark_background')
    plt.figure(figsize=(20,20))
    plt.scatter(0, 0, color='w')
    plt.plot(x - x0, y - y0, 'w')
    plt.axis('off');
    plt.title(planet, fontname='Humor Sans', fontsize=40)

    plt.savefig(f'{planet}.png')
```

```python
a_earth = 1
a_planet = dict_planets[planet]
# e_e = 0.0167
# e_v = 0.0067
P_earth = 1
P_planet = P_earth*a_planet**1.5
```

```python
P_synodic = P_earth*P_planet/np.abs(P_planet-P_earth)
```

```python
t = np.arange(0, P_synodic*12.39, 0.001)
```
