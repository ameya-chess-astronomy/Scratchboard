```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1.4,1.4,1000)
y = np.linspace(-1.4,1.4,1000)
```

```python
X,Y = np.meshgrid(x,y)
```

```python
a = 0.1
```

```python
U = (1-a)/np.sqrt(Y**2 + (X+a)**2) + a/np.sqrt(Y**2 + (X+a-1)**2) + 0.5*(X**2 + Y**2)
```

```python
Z=2*U
```

```python
U_ = (1-a)/np.sqrt((x+a)**2) + a/np.sqrt((x+a-1)**2) + 0.5*(x**2) 
```

```python
indL1 = np.where(U_== U_[600:800].min())[0][0]
indL2 = np.where(U_== U_[800:].min())[0][0]
indL3 = np.where(U_==U_[:400].min())[0][0]
earth = np.where(U_==U_[800:].max())
sun = np.where(U_==U_.max())
```

```python
ind_min = np.where(Z==Z.min())
```

```python
ind_max = np.where(Z==Z.max())
```
